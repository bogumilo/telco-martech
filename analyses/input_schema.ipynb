```python
import numpy as np
import pandas as pd
import xlrd
```

```python
# Load spreadsheet
xl = pd.ExcelFile('transformations.xls')

# Load a sheet into a dataframe by its name
df = xl.parse(xl.sheet_names[0])  # assuming we want to read the first sheet

# Select the first two columns
df = df.iloc[:, 1:3]

print(df.head(10))
```

```python
names = df.iloc[3:, :2].dropna().set_index("Unnamed: 1")
names = names.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
names = names.to_dict()["Unnamed: 2"].values()
names
```

```python
infile = pd.read_csv("input.csv", sep=",", skiprows=1, names=names)
infile
```

```python
infile.to_csv('input_schema.csv', index=False, header=True)
```
